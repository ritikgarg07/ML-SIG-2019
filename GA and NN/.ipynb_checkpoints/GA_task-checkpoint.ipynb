# Task 1: Extremizing A polynomial within an Interval
### The Following Code Returns a Polynomial Function given a set of coefficients

## Example:
$\left[ \begin{array} ,1&2&3 \end{array} \right]\rightarrow x^2 + 2x + 3 $


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def polynomial_creator(*coeffs):
    """ coefficients are in the form a_n, a_n_1, ... a_1, a_0
    """
    def polynomial(x):
        res = coeffs[0]
        for i in range(1, len(coeffs)):
            res = res * x + coeffs[i]
        return res
    return polynomial

def polynomial_plotter(polynomial, domain, num_points=100, show_plot = True):
    """ plots the polynomial in the given range with the give number of points
    """
    start, stop = domain
    points = np.linspace(start, stop,num_points)
    func_values = polynomial(points)
    plt.plot(points, func_values)
    if show_plot:
        plt.show()
```

```python
my_polynomial = polynomial_creator(3,-10,-150,3)
polynomial_plotter(my_polynomial, [1, 10])
```

### In a similar manner fill in the code below to return a multivariate function

## Example:
$\left[ \begin{array} ,1&2&3 \end{array} \right] , \left[ \begin{array} ,18&14&6&3 \end{array} \right] , \left[ \begin{array} ,12&3 \end{array} \right] \rightarrow f(x_1, x_2, x_3) = x_1^2 + 2x_1 + 3 + 18x_2^4 + 14x_2^3 + 6x_2^2 + 3x + 12x_3^2 + 3x $

### NOTE : The value of constant is given as the value of the first variable
Hint: You may use the polynomial creator function defined above


#### Optional/Bonus : Find a method to plot it
[How to plot multivariate functions](http://geog.uoregon.edu/bartlein/courses/geog495/lec05.html)


```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from random import randint


def multi_creator(*coefs):
    def multivariate_function(x):
        value = 0
        for ii,var in enumerate(x):
            temp_poly = polynomial_creator(*coefs[ii])
            value += temp_poly(var)
        return value
        
    return multivariate_function

def polynomial_plotter(polynomial, domain, num_points=100):
    start = domain[0]
    stop = domain[1]
    values = []
    X = []
    Y = []
    Z = []
    for ii in range(num_points):
        x = start + (stop-start)*randint(0,100)/num_points
        y = start + (stop-start)*randint(0,100)/num_points
        z = start + (stop-start)*randint(0,100)/num_points
        value = polynomial([x,y,z])
        values.append(value)
        X.append(x)
        Y.append(y)
        Z.append(z)
        
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    colors = ax.scatter(X, Y, Z, c=values, cmap=cm.viridis_r)
    fig.colorbar(colors)
    
    plt.show()
```

```python

my_polynomial = multi_creator([1, 2, 3], [18, 14, 6, 3], [12, 5])
polynomial_plotter(my_polynomial,[1,10])
```

# Now lets extremize it! (Using genetic algoritms)
Problem : Find the maxima of this function using genetic algos
For this we require
1. Fitness function
2. An encoder function
3. A mating function

## Notations
1. $n$ : The number of variables in the multivariate funtion
2. $m$ : Size of the population
3. $R_i$ : Range of the ith variable, as a tuple with max and min

## Fitness function
This is the function that our model will try to maximize. Since our goal is to maximize the multivariable polynomial, it will become our fitness function

## Encoder function
This functions takes in input genes and transforms its into meaningfull values of the variables.
#### Genes
For the purpose of this excercise, genes of the population are represented 2D numpy array of shape (n,m). **Our genes will only take values between -5 and 5**
#### Example
1. Polynomial $x_1 + x_2^2 + x_3 - x_3^3$
2. Population size :m=5
3. Ranges : (-3, 4), (-1, 2), (3, 9)
##### Input (Gene Pool)
$\left[ \begin{array}4.9&2.8&-3.4\\-3.5&1.4&4.5\\0.6&-3&1.8\\0.55&3.8&4.0\\-2& 2.6& 3 \end{array}\right]$
##### Output (Fitness Values)
$\left[ \begin{array} -52.413 & \\
-650.906 & \\
-346.734 & \\
-580.729 & \\
-466.014 & \\
-26.000 & \\
\end{array}\right]$

## Mating Funtion
This consists of :
1. Mutation Function - Input = Gene : Output = Slightly modified gene
2. Crossover function - Input = 2 genes : Output =  Crossovered genes

Of each gene pool only the top few are selected and mated to get the gene pool of the next iteration.

#### Note
The following illustration will introduce you to the basics of Genetic algorithms. We insist you to write a mutation function and a crossover function on your own. 
[This](https://towardsdatascience.com/genetic-algorithm-implementation-in-python-5ab67bb124a6) might inspire you.

```python
from random import randint, uniform
import copy

#parents pass on their genes, mutate and then are added to the population

#Global variables:
no_var = 3   #number of variables in fitness function, also equal to individual length
co_efs = [1, 2, 3], [18, 14, 6, 3], [12, 5]  #co-effs of variables 
fun = multi_creator(*co_efs)
fitness_fun = copy.deepcopy(fun)
pop_size = 5
l_bound = -5
r_bound = 5
parents_frac = 0.4
random_parent = 0.05  #prob of a random individual to become a parent
mutation_prob = 0.01
no_gen = 50   #no of generations
seed = [[0,0,0],[5,-5,0],[0,3.8,.78],[4.4,1.7,2.78],[3.9,-4.5,-1.2]]


def individual(l_bound, r_bound, no_var):
    return [uniform(l_bound, r_bound) for ii in range(no_var)]

def population(pop_size, l_bound, r_bound, no_var):
    return [individual(l_bound, r_bound, no_var) for ii in range(pop_size)]
    
def fitness(individual, fitness_fun):
    return fitness_fun(individual)

def select_parents(population, parents_frac, pop_size):
    sorted_fitness = [(fitness(individual, fitness_fun), individual) for individual in population]
    sorted_fitness = [tuple_ind[1] for tuple_ind in sorted(sorted_fitness)]
    #prints best fitness individual of generation
    print('individual: ' + str(sorted_fitness[0]))
    print(fitness(sorted_fitness[0], fitness_fun))
    print()
    parents = sorted_fitness[:int(pop_size*parents_frac)]  #looking for min fitness
    for individual in sorted_fitness[int(pop_size*parents_frac):]:
        if random_parent > uniform(0,1):
            parents.append(individual)
            
    return parents
    
def mutate(parents, mutation_prob, no_vars):
    for parent in parents:
        if mutation_prob > uniform(0,1):
            parent[randint(0,no_var-1)] *= uniform(0,1)
            parent[randint(0,no_var-1)] += uniform(0,1)  #in case 0 is a local extrema
    return parents
            
def crossover(parents, pop_size, no_vars):
    no_parents = len(parents)   #don't update no_parents to prevent children from becoming parents in the generation of their birth
    for ii in range(pop_size - no_parents):
        father = randint(0, no_parents - 1)
        mother = randint(0, no_parents - 1)
        child = [0 for var in range(no_var)]
        if father != mother:
            father = parents[father]
            mother = parents[mother]
            for jj in range(no_vars):
                if uniform(0,1) > 0.5:
                    child[jj] = father[jj]
                else: 
                    child[jj] = mother[jj]
        parents.append(child)
    return parents   #now a whole generation
def evolve(population):
    parents = select_parents(population, parents_frac, pop_size)
    parents = mutate(parents, mutation_prob, no_var)
    next_gen = crossover(parents, pop_size, no_var)
    return next_gen

def ga(seed, no_gen):
    next_gen = seed
    for ii in range(no_gen):
        print('Generation number: ' + str(ii+1))
        next_gen = evolve(next_gen)
```
